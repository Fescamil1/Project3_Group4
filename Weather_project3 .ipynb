```python
import pandas as pd
import datetime as dt
```

```python
df= pd.read_csv('Resources/WeatherEvents_Jan2016-Dec2020.csv', infer_datetime_format=True, parse_dates= ['StartTime(UTC)','EndTime(UTC)'])
df.head()
```

```python
#creating the Duration column 

df['Duration'] = (df['EndTime(UTC)'] - df['StartTime(UTC)']).dt.seconds
df.head()
```

```python
#converting the duration column into minutes 
df['Duration']= df['Duration'] / 60
df
```

```python
df = df.drop(columns= ['ZipCode', 'AirportCode', 'County' ])
df.head()
```

```python
dict = {"EventId" :"eventid",
        "Type":"type",
        "Severity":"severity",
        "StartTime(UTC)": "startTime",
        "EndTime(UTC)":"endTime",
        "TimeZone": "timeZone",
        "LocationLat": "lat",
        "LocationLng": "lng",
        "City":"city",
        "State": "state",
        "Duration": "duration"          
                
        
       }
  
# call rename () method
df.rename(columns=dict,
          inplace=True)
  
# print Data frame after rename columns
df
```

```python
#selecting the cities 
new_df = df.loc[df['state'].isin(['NY', 'CA', 'IL', 'TX', 'AZ','PA'])]
new_df = new_df.loc[new_df['city'].isin(['New York', 'Los Angeles', 'Chicago', 'Houston', 'Phoenix', 'Philadelphia', 'San Antonio', 'San Diego', 'Dallas','Austin'])]
new_df['year'] = pd.DatetimeIndex(new_df['startTime']).year
new_df
```

```python
stat_df = round(new_df.groupby(['year','city','type']).sum()['duration'] / 1440, 1)
stat_df.head()

stat_df = pd.DataFrame(stat_df)

stat_df = stat_df.reset_index()
stat_df['avg_perc_year'] = round(stat_df['duration'] / 365 * 100, 1)
stat_df
```

```python
city = list(stat_df['city'])
lat = []
lng = []

for i in range(len(city)):
    if str(city[i]) == "New York":
        lat.append(40.713)
        lng.append(-74.01)
    elif str(city[i]) == "Los Angeles":
        lat.append(34.05)
        lng.append(-118.24)
    elif str(city[i]) == "Chicago":
        lat.append(41.88)
        lng.append(-87.63)
    elif str(city[i]) == "Houston":
        lat.append(29.76)
        lng.append(-95.37)
    elif str(city[i]) == "Philadelphia":
        lat.append(39.95)
        lng.append(-75.17)
    elif str(city[i]) == "Phoenix":
        lat.append(33.45)
        lng.append(-112.07)
    elif str(city[i]) == "San Antonio":
        lat.append(29.42)
        lng.append(-98.49)
    elif str(city[i]) == "San Diego":
        lat.append(32.72)
        lng.append(-117.16)
    elif str(city[i]) == "Dallas":
        lat.append(32.78)
        lng.append(-96.80)
    elif str(city[i]) == "Austin":
        lat.append(30.27)
        lng.append(-97.74)   
        
stat_df['lat'] = lat
stat_df['lng'] = lng

stat_df
```
