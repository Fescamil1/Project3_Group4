```python
import pandas as pd
```

```python
df= pd.read_csv('Resources/WeatherEvents_Jan2016-Dec2020.csv', infer_datetime_format=True, parse_dates= ['StartTime(UTC)','EndTime(UTC)'])
df.head()
```

```python
df.dtypes
```

```python
#df['StartTime(UTC)'] = pd.to_datetime(df['Date'])
```

```python
#creating the Duration column 

df['Duration'] = (df['EndTime(UTC)'] - df['StartTime(UTC)']).dt.seconds
df.head()
```

```python
#converting the duration column into minutes 
df['Duration']= df['Duration'] / 60
df
```

```python
df = df.drop(columns= ['ZipCode', 'AirportCode', 'County' ])
df.head()
```

```python
dict = {"EventId" :"eventid",
        "Type":"type",
        "Severity":"severity",
        "StartTime(UTC)": "startTime",
        "EndTime(UTC)":"endTime",
        "TimeZone": "timeZone",
        "LocationLat": "lat",
        "LocationLng": "lng",
        "City":"city",
        "State": "state",
        "Duration": "duration"          
                
        
       }
  
# call rename () method
df.rename(columns=dict,
          inplace=True)
  
# print Data frame after rename columns
df
```

```python
#dependency to connect with sql alchelmy 
from sqlalchemy import create_engine
```

```python
#connecting
engine = create_engine('postgresql://postgres:postgres@localhost:5432/weather_db')
connection = engine.connect()
```

```python
#making sure of the connection!
engine.table_names()
```

```python
#changing the columns to lower case
new_df.columns= new_df.columns.str.lower()
new_df.head()
```

```python
new_df.to_sql(name='weatherhist', con=engine, if_exists='append', index=False)
```
